```python
import os
import pandas as pd, numpy as np
```

```python
def get_movie_data():
    
    unames = ['user_id','gender','age','occupation','zip']
    users = pd.read_table(os.path.join('../data','users.dat'), 
                          sep='::', header=None, names=unames)
    
    rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
    ratings = pd.read_table(os.path.join('../data', 'ratings.dat'), 
                            sep='::', header=None, names=rnames)
    
    mnames = ['movie_id', 'title','genres']
    movies = pd.read_table(os.path.join('../data', 'movies.dat'), 
                           sep='::', header=None, names=mnames)

    return users, ratings, movies
```

```python
users, ratings, movies = get_movie_data()
```

```python
""" Merge 'users' data with 'movie_ratings' data """

umr = users.merge(
    ratings.merge(movies.drop(columns='title').merge(
        movies.title.str.extract('(?P<title>.*)\((?P<year>[0-9]+)\)'),left_index=True, right_index=True)
                  ,left_on='movie_id',right_on='movie_id',how='left'),left_on='user_id',right_on='user_id',how='left')
umr
```

```python
umr = umr.merge(pd.DataFrame({'movie_id':[movie_id for movie_id,group in umr.groupby(['movie_id'])]
                              ,'ave_#_ratings':[len(group) for movie_id,group in umr.groupby(['movie_id'])]
                              ,'overall_rating':[np.mean(group['rating']) for movie_id,group in umr.groupby('movie_id')]}))
umr
```

```python
ave_num_movie_ratings = np.mean(umr['ave_#_ratings'])
ave_num_movie_ratings
```

```python
popular_movies = umr[(umr['ave_#_ratings'] > ave_num_movie_ratings)].sort_values('ave_#_ratings',ascending=False)
popular_movies
```
