```python
import os
import pandas as pd, numpy as np
```

```python
def get_movie_data():
    
    unames = ['user_id','gender','age','occupation','zip']
    users = pd.read_table(os.path.join('../data','users.dat'), 
                          sep='::', header=None, names=unames)
    
    rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
    ratings = pd.read_table(os.path.join('../data', 'ratings.dat'), 
                            sep='::', header=None, names=rnames)
    
    mnames = ['movie_id', 'title','genres']
    movies = pd.read_table(os.path.join('../data', 'movies.dat'), 
                           sep='::', header=None, names=mnames)

    return users, ratings, movies
```

```python
users, ratings, movies = get_movie_data()
```

```python
""" Merge 'ratings' data with 'movies' data """

movie_ratings = ratings.merge(movies.drop(columns='title').merge(movies.title.str.extract('(?P<title>.*)\((?P<year>[0-9]+)\)'), left_index=True, right_index=True), left_on='movie_id', right_on='movie_id', how='left')
print(movie_ratings)
```

```python
""" Merge 'users' data with 'movie_ratings' data"""

user_ratings = users.merge(movie_ratings, left_on='user_id', right_on='user_id', how='left')
print(user_ratings)
```

```python
""" Use groupby operation on 'user_ratings' data by utilizing 'title' and 'rating' as keys """

user_ratings_grouped = user_ratings.groupby(['title', 'rating'])
```

```python
""" Create a list that contains tuples with grouped data set keys and the number of a specific rating per movie """

rate_list = [(title, len(groups)) for title, groups in user_ratings_grouped]

for r in rate_list:
    print(r)
```

```python
""" Separate each movie into different categories """

five_star = sorted([movie for movie in rate_list if movie[0][1] == 5], key = lambda x:x[1], reverse=True)
four_star = sorted([movie for movie in rate_list if movie[0][1] == 4], key = lambda x:x[1], reverse=True)
three_star = sorted([movie for movie in rate_list if movie[0][1] == 3], key = lambda x:x[1], reverse=True)
two_star = sorted([movie for movie in rate_list if movie[0][1] == 2], key = lambda x:x[1], reverse=True)
one_star = sorted([movie for movie in rate_list if movie[0][1] == 1], key = lambda x:x[1], reverse=True)
```

```python
idx = 0; movies_total_stars = []

while idx < len(five_star):
    rating_sum = five_star[idx][1] + four_star[idx][1] + three_star[idx][1] + two_star[idx][1] + one_star[idx][1]
    movies_total_stars.append((five_star[idx][0][0],
                               rating_sum,
                               five_star[idx][1], four_star[idx][1], three_star[idx][1], two_star[idx][1], one_star[idx][1],
                               (5*five_star[idx][1] + 4*four_star[idx][1] + 3*three_star[idx][1] + 2*two_star[idx][1] + one_star[idx][1]) / rating_sum))
    idx += 1
```

```python
average_num_of_ratings = np.mean([movie[1] for movie in movies_total_stars])
```

```python
popular_movies = pd.DataFrame({'title':[movie[0] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              '# of ratings':[movie[1] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              '# of 5-stars':[movie[2] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              '# of 4-stars':[movie[3] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              '# of 3-stars':[movie[4] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              '# of 2-stars':[movie[5] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              '# of 1-star':[movie[6] for movie in movies_total_stars if movie[1] > average_num_of_ratings],
                              'average rating':[movie[7] for movie in movies_total_stars if movie[1] > average_num_of_ratings]})
```

```python
print(popular_movies)
```

```python
print(popular_movies.sort_values(by='average rating', ascending=False))
```
