```python
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=0)
t_df
```

**Use the groupby method to calculate the proportion of passengers that survived by sex.**

```python
for sc,group in t_df.groupby(['sex']):
    print("Number of",sc,"survivors, out of total survivors =",len(group[(group['survived'] > 0)]),":",
          len(t_df[(t_df['survived'] > 0)]))
```

**Calculate the same proportion, but by class and sex.**

```python
for sc,group in t_df.groupby(['sex','pclass']):
    print("Number of",sc[0],"survivors, of class",sc[1],", out of total survivors =",len(group[(group['survived'] > 0)]),":",
          len(t_df[(t_df['survived'] > 0)]))
```

**Children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex.**

*Note:* These categories do not include all the survivors because some survivors do not have a recorded age in the data.

```python
children = t_df[(t_df['age'] < 14)]
adolescents = t_df[(t_df['age'] >= 14) & (t_df['age'] < 21)]
adults = t_df[(t_df['age'] >= 21) & (t_df['age'] < 65)]
seniors = t_df[(t_df['age'] >= 65)]
```
